```python
df1 = df.groupby('규모').size().to_frame()
```

```python
df1
```

```python

filtered_df = df1[df1.index >= 7.7]
labels = list(filtered_df.index)
values = list(filtered_df[0])
```

```python
labels
```

```python
explode = [.1] * len(values)
colors=['r', 'g', 'b', 'c', 'm', 'y', 'orange', '#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
plt.pie(values, labels = labels, autopct='%.1f%%', startangle=90, counterclock=False, explode=explode, colors=colors)
plt.show() #'%.1f%%' 소수점 첫째 자리 % 포함 출력
```

```python
def custom_autopct(pct): return '{:.1f}%'.format(pct) if pct >=1.4 else '' #return ('%.1f%%' %pct) if pct >= 10 else ''
plt.pie(values, labels = labels, autopct=custom_autopct, startangle=90, counterclock=False, explode=explode, colors=colors,  pctdistance=0.7)
plt.title('전세계 규모7.7이상의 지진')
plt.legend(loc=(1.2, 0.3), title='지진규모')
plt.show() #'%.1f%%' 소수점 첫째 자리 % 포함 출력
```
