```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
font_path = "C:\\Users\\KOSMO\\AppData\\Local\\Microsoft\\Windows\\Fonts\\강원교육모두 Bold.ttf"
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name)
```

```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')
df
```

```python
max = df['키'].max()
max
min = df['키'].min()
min
```

```python
x = list(df['이름'])
x
y = list(df['키'])
y
colors=['r', 'g', 'b', 'k', 'c', 'm', 'y', 'orange']
bar = plt.bar(x, y, color = colors, alpha=0.5, width=0.9)
plt.ylim(min-10, max+10)
plt.xticks(rotation=105) #x 축의 이름 각도를 45도로 설정
plt.yticks(rotation=105) #y 축의 이름 각도를 45도로 설정
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
bar[3].set_hatch('\\')
bar[4].set_hatch('//')
bar[5].set_hatch('\\')
bar[6].set_hatch('/')
bar[7].set_hatch('.')
for idx, height in enumerate(y): plt.text(idx, height + 0.5, height, ha='center', color='k')
plt.show()
```

```python
colors=['r', 'g', 'b', 'k', 'c', 'm', 'y', 'orange']
plt.barh(x, y, color = colors, alpha=0.5)
plt.xlim(min-10, max+10)
plt.xticks(rotation=105) #x 축의 이름 각도를 45도로 설정
plt.yticks(rotation=105) #y 축의 이름 각도를 45도로 설정
plt.show()
```
