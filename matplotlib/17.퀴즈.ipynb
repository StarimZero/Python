```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
font_path = "C:\\Users\\KOSMO\\AppData\\Local\\Microsoft\\Windows\\Fonts\\강원교육모두 Bold.ttf"
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name)
data = {
 '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'], 
 '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015], 
 '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], #(단위 : 만 명) 
 '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
plt.bar(df['영화'], df['평점'])
plt.title('국내 관객수 Top8 영화 평점 정보')
plt.xlabel('영화제목')
plt.xticks(rotation=45)
plt.ylabel('평점')
```

```python
df_group = df.groupby('개봉 연도')[['관객 수','평점']].mean()
df_group
```

```python
plt.plot(df_group.index, df_group['평점'], marker='o', color = '#454')
plt.xticks([2006, 2012, 2013, 2014, 2015, 2017, 2019])
plt.ylim(7, 10)
```

```python
filt = df['평점'] >= 9.0
df[filt]
df[~filt]
values = [len(df[filt]), len(df[~filt])]
labels = ['9점 이상', '9점 미만']
wedgeprops={'width':0.8}
colors=['#DD291A', '#D29AB1', '#A10CBA', 'c', 'm', 'y', 'orange', '#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
plt.pie(values, labels=labels, autopct='%.2f%%', colors=colors, counterclock=False, startangle=90, wedgeprops=wedgeprops)
plt.legend(loc=(1, 0))
plt.show()
```
