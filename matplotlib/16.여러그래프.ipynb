```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
font_path = "C:\\Users\\KOSMO\\AppData\\Local\\Microsoft\\Windows\\Fonts\\강원교육모두 Bold.ttf"
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name)
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10)) #2x2에 해당하는 plot 생성
fig.suptitle('여러 그래프 넣기')
#그래프1
a1 = axs[0,0]
a1.bar(df['이름'], df['키'], label='키') #데이터 설정
a1.set_title('막대 그래프') #제목
a1.legend() #범례
a1.set(xlabel='이름', ylabel='키', ylim=(165, 210)) #x, y 축 label
a1.set_facecolor('#07204d') #전면 색
a1.grid(linestyle=':', linewidth=1.5)
#그래프2
a2=axs[0,1]
a2.plot(df['이름'], df['수학'], label='수학')
a2.plot(df['이름'], df['영어'], label='영어')
a2.plot(df['이름'], df['사회'], label='사회')
a2.legend()
#그래프3
axs[1, 0].barh(df['이름'], df['키'])
#그래프4
axs[1, 1].scatter(df['영어'], df['사회'], color='green', alpha=0.5)

```
