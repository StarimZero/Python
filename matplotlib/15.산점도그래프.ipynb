```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
font_path = "C:\\Users\\KOSMO\\AppData\\Local\\Microsoft\\Windows\\Fonts\\강원교육모두 Bold.ttf"
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name)
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')
df
```

```python
x = list(df['영어'])
x
```

```python
y = list(df['수학'])
y
```

```python
df.index
```

```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel('영어점수')
plt.ylabel('수학점수')
for idx, no in enumerate(df.index):
    plt.text(x[idx], y[idx]+2, no)
plt.show()
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000
sizes
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
sizes = df['학년'] * 100 #1학년=500, 2학년=1000, 3학년=1500
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='inferno', alpha=.8)
plt.xlabel('영어 점수') #cmap='plasma', cmap='inferno', cmap='magma', cmap='cividis'
plt.ylabel('수학 점수')
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='inferno', alpha=1) #alpha 겹칠 경우
for idx, no in enumerate(df.이름):
    plt.text(x[idx], y[idx]+2, no)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar(ticks=[1, 2, 3], label='학년', shrink=0.5, orientation='horizontal')
```
