```python
dataset = pd.read_csv('data/LinearRegressionData.csv')
dataset
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X, len(X)
```

```python
reg.fit(X, y)
```

```python
y_train, len(y_train)
```

```python
y_test, len(y_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, reg.predict(X_train), color='green')
plt.title('Score by hours (train data)')
plt.xlabel('시간')
plt.ylabel('점수')
plt.show()
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_train, reg.predict(X_train), color='green')
plt.title('Score by hours (test data)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
reg.coef_, m
```

```python
reg.intercept_, b
```

```python
reg.score(X_test, y_test)
```

```python
reg.score(X_train, y_train)
```

```python
from sklearn.linear_model import SGDRegressor
sr = SGDRegressor()
sr.fit(X_train, y_train)
```

```python
plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, sr.predict(X_train), color='green')
plt.title('SGD')
plt.xlabel('시간')
plt.ylabel('점수')
plt.show() #LinearRegression이용한 직선과 거의 유사하다.
```

```python
sr.coef_, sr.intercept_
```

```python
reg.coef_, m, reg.intercept_, b
```

```python
sr.score(X_test, y_test)
```

```python
sr.score(X_train, y_train), reg.score(X_train, y_train)
```

```python
from sklearn.linear_model import SGDRegressor
sr = SGDRegressor(max_iter=10000, eta0=1e-3, random_state=0, verbose=1)
sr.fit(X_train, y_train)
```

```python
plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, sr.predict(X_train), color='green')
plt.title('Score by hours (test data: SGD)')
plt.xlabel('시간')
plt.ylabel('점수')
plt.show()
```
