```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('data/KMeansData.csv')
dataset
```

```python
X = dataset.iloc[:,:].values # X = dataset.values와 결과가 같다.
X
```

```python
plt.scatter(X[:, 0], X[:, 1]) # x축:hour, y축: score 
plt.title('Scores by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.xlim(0, 100)
plt.xlim(0, 100)
plt.title('Scores by hours')
plt.xlabel('hour')
plt.ylabel('score')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
X
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(X[:, 0], X[:, 1])
plt.title('Scores by hours')
plt.xlabel('hour')
plt.ylabel('score')
plt.show()
```

```python
from sklearn.cluster import KMeans
inertia_list = [] #cluster에 속한 점들이 얼마나 가깝게 모여 있는지를 나타내는 값들
for i in range(1, 11):
 #중심점(centroid)을 K-menas++ 방식으로 계산하여 객체를 생성한다. (n_init=10 : 디폴트 수)
 kmenas = KMeans(n_clusters=i, init='k-means++', random_state=0, n_init=10)
 kmenas.fit(X) #학습 후 모델을 생성한다.
 inertia_list.append(kmenas.inertia_) #각 지점으로부터 중심점(centroid)까지의 거리 제곱의 합(inertia)
plt.plot(range(1, 11), inertia_list)
plt.title('Elow Method')
plt.xlabel('n_cluster')
plt.ylabel('inertia')
plt.show()
```

```python
K = 4
kmeans = KMeans(n_clusters=4, random_state=0, n_init=5)
y_kmeans = kmeans.fit_predict(X) #X의 점들이 어느 cluster에 속하는지 예측해 본다. 
y_kmeans #X의 점들의 어느 cluster에 속하는지 출력해 본다.
```

```python
dataset
```

```python
dataset['cluster'] = y_kmeans
dataset
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
plt.figure(figsize=(7, 7))
for i in range(K):
    filt = y_kmeans == i
    plt.scatter(X[filt, 0], X[filt, 1], s=100, edgecolor='black')
    plt.scatter(centers[i, 0], centers[i, 1], s=300, edgecolor='black', color='#d8ff99', marker='d')
    plt.text(centers[i, 0], centers[i, 1], i, va='center', ha='center')
plt.title('Score / Hours')
plt.xlabel('hour')
plt.ylabel('score')
plt.show()
```

```python
X_org = sc.inverse_transform(X)
X_org[:5]
```

```python
centers_org = sc.inverse_transform(centers)
centers_org
```

```python
plt.figure(figsize=(7, 7))
for i in range(K):
    filt = y_kmeans == i
    plt.scatter(X_org[filt, 0], X_org[filt, 1], s=100, edgecolor='black')
    plt.scatter(centers_org[i, 0], centers_org[i, 1], s=300, edgecolor='black', color='#d8ff99', marker='h')
    plt.text(centers_org[i, 0], centers_org[i, 1], i, va='center', ha='center')
plt.title('Score / Hours')
plt.xlabel('hour')
plt.ylabel('score')
plt.show()
```
